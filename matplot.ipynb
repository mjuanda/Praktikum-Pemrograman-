```python
import matplotlib.gridspec as GridSpec

np.random.seed(1)
x = np.arange(21)
y = 2.5 + 1.65 * x
yn = y + np.random.rand(len(x))

t = np.linspace(0,5,501)
y1 = np.sin(2*np.pi*t)
y2 = 1.6 * np.cos(np.pi*t)

x2 = np.linspace(-2,2,201)
yl = -2.5 + 1.65 * x2
yq = -1.5 + 2*x2 - 2.4*x2**2
yln = 1 + np.log(np.abs(1+x2))  

ti = np.linspace(0,5,501)
y11 = np.sin(2*np.pi*ti) + 1.2 *np.sin(2.5*np.pi*ti - 1)
y22 = 1.6 * np.cos(1.6*np.pi*ti + 2.3)

fig , ax = plt.subplots(2,2, figsize = (10,10))



## [0][0]
ax[0][0].scatter(x,yn,label= "Data", c = "b", marker= "*", s= 100)
ax[0][0].plot(x,y,label= "fitting data", c = "r")
ax[0][0].set_title("[0]")
ax[0][0].set_ylabel("y(t)")
ax[0][0].set_xlabel("x")
ax[0][0].set_ylabel('y')
ax[0][0].set_xlabel('x')
ax[0][0].legend()
ax[0][0].grid()

## [0][1]
ax[0][1].plot(t,y1, label= "sin (2t)", c = "magenta", linewidth = 2)
ax[0][1].plot(t,y2, label= "1.6cos(t)", c = "tab:green", linestyle = "--")
ax[0][1].set_title("[1]")
ax[0][1].set_ylabel("y(t)")
ax[0][1].set_xlabel("x")
ax[0][1].grid()
ax[0][1].legend()

## [1][0]
ax[1][0].plot(x2,yl, c="r", label = "y = -2.5 + 1.65x",linewidth = 2)
ax[1][0].plot(x2,yq, linestyle = "--", c = "b", label = "y = 1.5 + 2x - 2.4x^2",linewidth = 2)
ax[1][0].plot(x2,yln, linestyle = "-.", c='k', label = "y= 1 + ln(|1 + x|)", linewidth = 2)
ax[1][0].set_xlabel("x")
ax[1][0].set_ylabel("y")
ax[1][0].grid()
ax[1][0].legend()


## [1][1]
ax[1][1].plot(ti, y11,linewidth = 2, label = "sin(2t) + 1.2sin(2.5nt - 1)", c = "r")
ax[1][1].plot(ti, y22, linestyle= ":",linewidth = 2, label = "1.6cos(1.6t+2.3)", c= "magenta")
ax[1][1].set_ylabel('y')
ax[1][1].set_xlabel('x')
ax[1][1].grid()
ax[1][1].legend(loc = 3)
```
